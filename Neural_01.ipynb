```
def f_v_gnoise(v, dt,I_dc,I_syn): 
    return (0.04 * v * v + 5 * v + 140 - u + I_dc+I_syn) *dt

def f_u(u,dt): 
    return a * (b * v - u) *dt

def solve_diff_gnoise(dt,u,v,I_dc,I_syn): 
    dv = f_v_gnoise(v, dt,I_dc,I_syn) 
    du = f_u(u,dt) 
    v += dv
    u += du
    return u,v,dv

def step_gnoise(dt,u,v, I_dc,I_syn):
  if v >= thrs:
    v = c 
    u += d
    dv = 0
  else: 
    u,v,dv = solve_diff_gnoise(dt,u,v,I_dc,I_syn)
  return u,v,dv

def sig_syn(g,vi,ti,tf):
  V0 = 0
  Tau_s = 1.7
  Tau_f = 0.2
  Ts = (ti-tf)/Tau_s
  Tf = (ti-tf)/Tau_f
  sig = (g*(np.exp(-Ts)-np.exp(-Tf))*(V0-vi)/(Tau_f-Tau_s))
  return sig
```
